## Formula used in Simple Linear Regression to calculate slope (m) and intercept (c).

We fit the line:

$[
y = m x + c
]$

---

### Slope (m) formula
- Slope (m) = covariance(x, y) / variance(x)

$[
m = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{\sum (x_i - \bar{x})^2}
]$


---

### Intercept (c) formula
- Intercept (c) = mean(y) − m × mean(x)

$[
c = \bar{y} - m \bar{x}
]$

Where:

- $( \bar{x} )$ = mean of x
- $( \bar{y} )$ = mean of y







```python
class MySLR:

    def __init__(self):
        self.m = None
        self.c = None

    def fit(self, X_train, y_train):

        num = 0
        den = 0

        for i in range(X_train.shape[0]):
            num = num + (X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean())
            den = den + (X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean())

        self.m = num / den
        self.c = y_train.mean() - (self.m * X_train.mean())

        print(self.m)
        print(self.c)
            

    def predict(self, X_test):
        return self.m * X_test + self.c
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"placement.csv")
df.head()
```

```python
sns.scatterplot(data = df, x = "cgpa", y = "package")
plt.xlabel("CGPA")
plt.ylabel("Package(in lpa)")
```

```python
X = df.iloc[:, 0].values # Input Feature
y = df.iloc[:, -1].values # Output Feature
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=2)
```

```python
X_train.shape
```

```python
# Model Building
lr = MySLR()
lr.fit(X_train, y_train) 
```

```python
lr.predict(X_test[0])
```

```python
lr.m # Slope
```

```python
lr.c # Intercept
```

```python
sns.scatterplot(data = df, x = "cgpa", y = "package")
plt.plot(X_train, lr.predict(X_train), color = "red")
plt.xlabel("CGPA")
plt.ylabel("Package(in lpa)")
```

```python
y_pred = lr.predict(X_test) # Prediction
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error, r2_score
```

```python
# Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print("Mean Absolute Error :", mae)
print("Mean Sqaured Error :", mse)
print("Root Mean Sqaured Error :", rmse)
print("R2 Score:", r2)
```
